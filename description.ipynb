## Overview

In this assignment, you will build a RESTful API using **Flask-RESTX** that supports two types of users — **Admin** and **User** — interacting with movie data.

- The **Admin** can create and manage movies using real-world data imported from CSV files.
- **Users** can browse and search movies.
- There is a single dedicated `admin` account that can create and delete User accounts.

Your API must:

- Store data in a local **SQLite** database.
- Expose endpoints that are fully documented and testable via **Swagger UI**.

### Requirement Set 1: User Management and Access Control (5 marks)

The API must support two account types: **Admin** and **User**.

- There is only one dedicated **Admin** account with username `admin`. There is no limit to the number of **User** accounts. 
- The `admin` account can access all endpoints.
- The `admin` can create and delete User accounts.
- The `admin` can activate or deactivate User accounts at any time. API access is available only to active accounts. Deactivated accounts lose their access to the API unless they are reactivated.
- The `admin` can retrieve a list of all accounts and their full details.

Your system must automatically create the following accounts on first run:

| Username | Password | Role  |
|----------|----------|-------|
| admin    | admin    | Admin |
| user     | user     | User  |

Additional constraints:

- The Admin can perform all actions available in the system.
- Users can perform all actions **except**:
  - User management
  - Importing movies

### Requirement Set 2: Importing Movies (2 marks)

Only the `admin` account can import movie and credit data.

- The Admin must be able to import two CSV files via the API. The files must be uploaded as file attachments using the multipart/form-data Content-Type.
  - [credits.csv](`credits.csv`)
  - [movies.csv](`movies.csv`)
- The imported data must be stored in the local SQLite database.
- You must process the provided CSV files and extract any information required to support the remaining requirement sets.
- Re-importing data must replace existing records in the database.
- You may assume that the dataset schema is consistent with the provided sample files. However, your API must validate uploaded files and return clear, appropriate error messages if the files are invalid, malformed, or do not conform to the expected schema.

### Requirement Set 3: Exploring Movies (3 marks)

All accounts must be able to:

- Retrieve detailed information about a specific:
  - Movie (by ID)
  - Crew member (by ID)
  - Cast member (by ID)

All accounts must also be able to search for movies by:

- Title (e.g., `Pirates of the Caribbean: At World's End!`)
- Crew member name (e.g., `Gore Verbinski`)
- Cast member name (e.g., `Johnny Depp`)

Search requirements:

- Case-insensitive (e.g., `avatar`)
- Partial matching supported (e.g., `pirates of`, `depp`, `Verbinski`)
- Use the **RapidFuzz** library
- The search may return zero or more results

---

### Requirement Set 4: Playlists (3 marks)

All accounts must be able to manage private playlists.

Users can:

- Create unlimited private playlists (A playlist is a collection of movies and is identified by a unique name. However, different users may have playlists with the same name.)
- Delete their playlists
- Add movies to playlists
- Remove movies from playlists
- View their list of playlists
- View all movies within a playlist
- Persist data in your local database

> Playlists must be private and accessible only to the owner.

### Requirement Set 5: Security and Monitoring (4 marks)

The purpose of the following requirement is to provide **operational insights into API usage from both a user and security perspective**.

You are tasked to help users and administrators quickly understand patterns such as:

- **Daily API usage trends** for an account (e.g., number of requests per day)
- **Relative activity levels** (e.g., identifying heavy API usage)

From a business and security standpoint, these statistics help to:

- Monitor how frequently the API endpoints are being used 
- Detect unusual or potentially abusive access patterns in endpoint level
- Provide transparency to account holders about their own API consumption


Your ouput must:

- Clearly communicate the usage statistics
- Be presented in the following formats based on the user's request:
  - As a downloadable CSV file
  - As a visualisation image (e.g., using `matplotlib`) that can be viewed in a browser

> All usage data must be persisted in a local database
> 
> The **Admin role has:
> - Access to usage statistics for all accounts (including both Admin and User accounts)
> - Access to aggregated statistics across all users, excluding Admin-specific activity
>  
> User accounts:
> - Can only access their own API usage statistics
>
> All timestamps should use the Sydney time zone (AEST/AEDT), regardless of the user’s local system time.
